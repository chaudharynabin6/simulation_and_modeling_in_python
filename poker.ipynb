```python
# calculating observed frequency
import math
observed_freq = []
for i in range(10):
    observed_freq.append(0)

N = total_numbers =  len(inputs)
n = n_class = 10

print(f"observed_freq before:{observed_freq} ")
for input in inputs:
    index = input*N/n
    index = math.floor(index)
    index = int(index)
    observed_freq[index] += 1
total = 0
for i in observed_freq:
    total = total + i
print(f"observed_freq after : {observed_freq} ")
print(f"total : {total} ")


```

```python
# calculating expected frequency
expected_freq = []

for i in range(10):
    expected_freq.append(10)

# calculating oi - ei

oi_minus_ei = []
oi_minus_ei_sq = []
oi_minus_ei_sq_div = []
for (oi,ei) in zip(observed_freq,expected_freq):
    # oi - ei of each 
    value = oi - ei
    oi_minus_ei.append(value)
    
    # oi - ei square of each
    sq_value = value * value
    oi_minus_ei_sq.append(sq_value)

    # oi - ei square div by ei
    div_sq_value = sq_value / ei
    oi_minus_ei_sq_div.append(div_sq_value)
    


print(f" observed frequency : {observed_freq} ")
print(f" expected frequency : {expected_freq} ")
print(f" oi_minus_ei  : {oi_minus_ei} ")
print(f" sq value  : {oi_minus_ei_sq} ")
print(f" sq div value  : {oi_minus_ei_sq_div} ")



```
